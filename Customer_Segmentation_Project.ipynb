```python
import seaborn as sns
import matplotlib.pyplot as plt

# Age distribution
sns.histplot(df['age'], bins=10, kde=True)
plt.show()

# Annual Income distribution
sns.histplot(df['income'], bins=10, kde=True)
plt.show()

# Spending Score distribution
sns.histplot(df['spending_score'], bins=10, kde=True)
plt.show()
```

```python
df.isnull().sum()
```

```python
X = df[['income', 'spending_score']]  # We will cluster based on Income and Spending
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))
sns.scatterplot(x='income', y='spending_score', hue='Cluster', palette='Set1', data=df, s=100)
plt.title('Customer Segmentation')
plt.show()
```

# **Insights and Conclusions**

After performing K-Means clustering on customer data, the following segments were identified:

- **Cluster 0:** High income, high spending → VIP customers; they are most valuable for premium offers.  
- **Cluster 1:** Low income, low spending → Budget-conscious customers; potential for loyalty programs.  
- **Cluster 2:** High income, low spending → Potential targeted offers; can be encouraged to spend more.  
- **Cluster 3:** Medium income, medium spending → Regular customers; maintain engagement.  
- **Cluster 4:** Low income, high spending → Occasional high spenders; analyze behavior for targeted deals.  

**Business Value:**  
- Helps businesses **tailor marketing strategies** per customer segment.
